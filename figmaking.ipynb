```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib qt

shapes = []

plt.figure()
plt.xlim(-100, 1100)
plt.ylim(-300, 10)
plt.grid(True)
plt.title("click to draw.")

while True:
    print("Press enter to continue to the next shape")
    
    points=plt.ginput(n=-1, timeout=-1) 
    
    if len(points)==0:

        break

    points_array=np.array(points)
    points_array[:, 0] = np.clip(points_array[:, 0], 0, 1000)   
    points_array[:, 1] = np.clip(points_array[:, 1], -250, 0)  
    prec_x = 50.0
    prec_y = 2.0
    points_array[:, 0] = np.round(points_array[:, 0] / prec_x) * prec_x
    points_array[:, 1] = np.round(points_array[:, 1] / prec_y) * prec_y
    shapes.append(points_array)

    plt.plot(points_array[:,0], points_array[:,1], '-o')
    plt.draw()
    
    ans = input("enter =new shape , q=quit")
    if ans.lower()=='q':
        break

for i, s in enumerate(shapes):
    filename = f"shape_{i}.txt"
    np.savetxt(filename, s)
    print(f"filesaved as: {filename}")

plt.show()
```
